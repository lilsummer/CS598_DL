## Plot figures

### Loss Curve 2
* Corresponding to Version Number: 26
* Crop center for both train and valid
* 20 epochs

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('../cg_experiment/resnet_05e_20_baseline.txt', 'rb') as fb:
    loss = pickle.load(fb)
```

```python
loss[:10]
```

```python
train_loss, val_loss = [], []
for i,v in enumerate(loss):
    if i%2==0:
        train_loss.append(v)
    else:
        val_loss.append(v)
plt.figure(figsize=(8,5))
plt.style.use('seaborn-poster')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(np.arange(20), train_loss, label='train')
plt.plot(np.arange(20), val_loss, label='valid')
plt.legend()
plt.tight_layout()
plt.savefig('loss_curve_resnet_20epochs_new_baseline.pdf')
```

```python
import pandas as pd
```

#### ResNet

```python
baseline = pd.read_csv('../cg_experiment/pc_curve_res_base.csv')
crop = pd.read_csv('../cg_experiment/pc_curve_res_crop.csv')
distor = pd.read_csv('../cg_experiment/pc_curve_res_distort.csv')
zoom = pd.read_csv('../cg_experiment/pc_curve_res_zoom.csv')
```

```python
plt.figure(figsize=(8,5))
plt.plot(baseline.recall, baseline.precision, label='base', linewidth=1)
plt.plot(zoom.recall, zoom.precision, label='zoom',linewidth=1)
plt.plot(crop.recall, crop.precision, label='crop', linewidth=1)
plt.plot(distor.recall, distor.precision, label='distortion', linewidth=1)
plt.legend()
plt.ylabel('Precision')
plt.xlabel('Recall')
```
